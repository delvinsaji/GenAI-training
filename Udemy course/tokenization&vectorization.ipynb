```python
paragraph = """
Dr. APJ Abdul Kalam is a famous name in the whole world. He is counted among the greatest scientists of the 21st century. Even more, he becomes the 11th president of India and served his country. He was the most valued person of the country as his contribution as a scientist and as a president is beyond compare. Apart from that, his contribution to the ISRO (Indian Space Research Organization) is remarkable. He headed many projects that contributed to the society also he was the one who helped in the development of Agni and Prithvi missiles. For his involvement in the Nuclear power in India, he was known as “Missile Man of India”. And due to his contribution to the country, the government awarded him with the highest civilian award.
"""
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
sentence_list = []
for sentence in sentences:
    word_list = word_tokenize(sentence)
    word_list = [lemmatizer.lemmatize(word,pos = 'v') for word in word_list if word not in set(stopwords.words('english'))]
    sentence_list.append(' '.join(word_list))
```

```python
import pandas as pd
messages = pd.read_csv("C:\OneDrive\OneDrive - Tredence\Attachments\GenAI\Course\spam.csv",encoding="ISO-8859-1")
```

```python
messages.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis = 1,inplace = True)
messages.columns = ["ham/spam","message"]

messages["spam"] = messages["ham/spam"].apply(lambda x: 1 if x == "spam" else 0)
messages.drop("ham/spam",axis = 1,inplace = True)
```

```python
messages.head()
```

```python
def text_preprocessing(sentence):
    sentence = sentence.lower()
    
    lematizer = WordNetLemmatizer()
    sentence_list = wordpunct_tokenize(sentence)

    final = []
    for items in sentence_list:
        final.append(lematizer.lemmatize(items,pos='v'))
    
    final = [word for word in final if word not in set(stopwords.words("english"))]
    return " ".join(final)

messages["processed_text"] = messages["message"].apply(text_preprocessing)
```

```python
messages.head()
```

```python
corpus = messages["processed_text"].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
X = cv.fit_transform(corpus).toarray()
```
